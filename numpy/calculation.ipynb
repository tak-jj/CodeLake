# **Calculation**

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 행렬 비교
arr1 = np.array([1, 2, 3, 4, 5, 6, 7])
arr2 = np.array([1, 2, 3, 10])

print(np.isin(arr1, arr2))
print(np.isin(arr2, arr1))
# 행렬의 각 위치 값을 비교하여 boolean을 반환한다. // compare values in array in same position, return boolean.
```

## array calculation

```python
a1 = np.array([[1, 2], [3, 4]])
a2 = np.array([[5, 6], [7, 8]])
```

```python
print(a1 + a2) # np.add(a1, a2)
print(a1 - a2) # np.subtract(a1, a2)
print(a1 * a2) # np.multiply(a1, a2)
print(a1 / a2) # np.divide(a1, a2)
```

```python
print(np.dot(a1, a2))
# 행렬곱 // matrix multiply
```

```python
# broadcast
# 형태가 같지 않은 배열도 계산할 수 있다.
# can calculate arrays in different shape.
a1 = np.array([1, 2 ,3])
a2 = 3

print(a1 * a2)
# 둘 중 하나가 1차원 또는 스칼라 값일 경우 계산할 수 있다.
# if one of two is 1-dimension array or scalar value, can calculate.
```

```python
a1 = np.array([1, 2, 3])
a2 = np.array([[1], [2], [3]])

print(a1.shape)
print(a2.shape)
print(a1 * a2)
# 차원의 축이 일치할 경우 계산할 수 있다.
# if axis match, can calculate.
```

```python
# np.sum()
# 배열 원소의 합
# sum of array elements
a = np.arange(10)

print(np.sum(a))
```

## math function

https://numpy.org/doc/stable/reference/routines.math.html

```python
# np.sin(), np.cos(), np.tan()
# 삼각함수
# trigonometric function
x = np.linspace(0, 7, 100)
y1 = np.sin(x)
y2 = np.cos(x)

plt.figure(figsize=(9,3))
plt.plot(x, y1)
plt.plot(x, y2)
plt.show()
```

```python
# np.sqrt()
# 제곱근
# root
x = np.sqrt(2)
print(x)
```

```python
# np.exp()
# e의 지수
# exponential of e
x = np.exp(2)
print(x)
```

```python
# np.log()
# 자연로그
# natural log
x = np.log(2)
print(x)
```

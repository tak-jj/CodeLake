# **meshgrid**

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# np.meshgrid()
# 격자형 데이터를 생성한다.
# create grid data.
a = np.linspace(1, 10, 10)
b = np.linspace(11, 20, 10)
x, y = np.meshgrid(a, b)
print(x)
print(y)
```

```python
plt.scatter(x, y)
plt.grid()
plt.show()
```
